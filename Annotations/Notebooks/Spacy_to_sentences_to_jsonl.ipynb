```python
phrase_list = []
for filepath in glob.iglob(r'.\*new.txt'):
    print(filepath)
    txt = filepath.strip('.\\')
    text_file = open(txt, 'r', encoding='utf-8')
    text = text_file.read()
    nlp = spacy.load('nb_core_news_sm', exclude=["parser"])
    nlp.enable_pipe("senter")
    doc = nlp(text)
    prev_sent = 'None'
    for sent in doc.sents:
        sentence = sent.text.lstrip().replace('\n', '')
        if len(sentence) > 100:
            phrase_list.append(sentence)
         #   print(this_sent, len(this_sent))
        elif len(sentence) < 100 and prev_sent != 'None':
            this_sent = prev_sent + sentence
            if len(this_sent) > 100:
                prev_sent = this_sent + '. '
            else:
                prev_sent = 'None'
                phrase_list.append(sentence)
                #print(this_sent, len(this_sent))
        elif len(sentence) < 100 and prev_sent == 'None':
            prev_sent = sentence + '. '
        
                

```

```python
random_list = phrase_list
random.shuffle(random_list)
```

```python
def list_to_jsonl(this_list, json_name):
    with jsonlines.open(json_name, mode='w') as writer:
        for line in this_list:
            file_lines = line
            text = json.dumps({'text': file_lines})
            text = json.loads(text)
            print(text)
            writer.write(text)
```
