```python
def pdf_2_txt(pdf, file):
    file_txt = open(file, 'a', encoding='utf-8')
    with pdfplumber.open(pdf) as pdf:
        for page in range(len(pdf.pages)):
            this_page = pdf.pages[page]
            if this_page.extract_text() != None:
                file_txt.write(this_page.extract_text())
    file_txt.close()
    clean_files(file)
```

```python
def clean_files(filename):
    f = str(filename)
    with open(filename, 'r', encoding='utf-8') as filename:
        new_filename = f.replace('.txt', '_new.txt')
        new_file = open(new_filename, 'a', encoding='utf-8')
        file_lines = filename.readlines()
        for text_line in enumerate(file_lines):
            text = text_line[1].replace('\n', '')
            text = text.replace('➦', '')
            text = text.lstrip()
            if len(text) > 1 and text != ' ':
                new_file.write(text + '\n')
        new_file.close()
        
```

```python
import jsonlines
import json
import pdfplumber
import glob
for filepath in glob.iglob(r'.\*_new.txt'):
    filepath = filepath.strip('.\\')
    f = str(filepath)
    with open(filepath, 'r', encoding='utf-8') as filename:
        new_filename = f.replace('_new.txt', '.jsonl')
        with jsonlines.open(new_filename, mode='w') as writer:
            file_lines = filename.readlines()
            for text_line in enumerate(file_lines):
                text = text_line[1].replace('\n', '')
                text = json.dumps({'text': text})

                text = json.loads(text)
                #if len(text) > 1:
                print(text)
                writer.write(text)
```
