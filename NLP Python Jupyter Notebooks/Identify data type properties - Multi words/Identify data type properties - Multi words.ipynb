```python
#
# Get the stop words
#

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

forward_english_stopwords = set(stopwords.words('english'))

# Legger til alt som ikkje er til hjelp for å finne dobbeltord
forward_english_stopwords.update(['.',',','(',')','c',';','..',':','b','c','the','in',"'s",'no','an','etc','where','-'])

# Removing the months
forward_english_stopwords.update(['january','february','march','april','may','june','july','august','september','october','november','december'])

forward_english_stopwords
```

```python
# Does the input string contain a digit?
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def hasSlash(inputString):
    return any(char == '/' for char in inputString)
```

```python
#
# Print bigram with no stopwords
#

from nltk import ngrams

for sentence_index_number,sentence_as_a_string in enumerate(forward_sent_tokens):
    token = nltk.word_tokenize(sentence_as_a_string)
    bigram_list = list(ngrams(token, 2))
    
    for bigram in bigram_list:
        
        if hasNumbers(bigram[0]) or hasNumbers(bigram[1]):
            continue # skip if containing numbers
        elif hasSlash(bigram[0]) or hasSlash(bigram[1]):
            continue # skip if containing /
        
        bigram_in_a_list = [bigram[0],bigram[1]]
        
        bigram_without_stopword = [bigram_split for bigram_split in bigram_in_a_list if bigram_split.lower() not in forward_english_stopwords]
        
        if len(bigram_without_stopword) == 2:
            print(bigram_without_stopword)
        
```

```python
#
# fjern bøying
#

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''

lines_with_lemmatized_words_and_tag = []
    
for token_line in lines_with_word_type:
    
    lemmatizer = WordNetLemmatizer()
    lemmatized_single_words_with_tag = []

    for (word, pos) in token_line:
        wordnet_pos = get_wordnet_pos(pos)
        if wordnet_pos != '':
            
            # Auto lemmatizer
            lemmatized_word = lemmatizer.lemmatize(word.lower(), wordnet_pos)
            
            # Custom lemmatizer
            if lemmatized_word == 'metres':
                lemmatized_word = 'metre' # (Commonwealth spelling)
            elif lemmatized_word == 'meter': # (American spelling)
                lemmatized_word = 'metre'
            
            lemmatized_single_words_with_tag.append((lemmatized_word, pos))
    
    lines_with_lemmatized_words_and_tag.append(lemmatized_single_words_with_tag)
    
lines_with_lemmatized_words_and_tag
```
