```python
#
# Lastar inn teksten vi jobbar med
#

with open('./Input text - Chapter 3.txt', 'r') as f:
    file_contents = f.read()
    
print(file_contents)
```

```python
#
# Splitte opp linjehopp frå tekst fil.
# Eksempel: Fire, machinery and electric installations\nSection 7.
#

sentences = file_contents.split('\n')
sentences
```

```python
# Transform list to one text
one_text = " ".join(sentences)
one_text
```

```python
from nltk.tokenize import word_tokenize

word_tokens = word_tokenize(one_text, language="english")
word_tokens
```

```python
from nltk.tokenize.punkt import PunktSentenceTokenizer

pst = PunktSentenceTokenizer()

sentences = pst.sentences_from_tokens(word_tokens)
list(sentences)
```

```python
# Splitte opp tekst etter setningar.
# Det gir ikkje meining å smelete setningar saman når ein leiter etter ord som høyrer saman.
# Vi leitar derfor etter ord som høyrer saman innanfor ein setning.

sent_tokens = pst.tokenize(one_text)

sent_tokens
```

# Remove all the words we don't need

```python
#
# Clean the lines up
#

import re

lines = []

for line in sent_tokens:
    
    # remove all numbers
    edit_1 = re.sub(r"\d","", line)
    
    edit_2 = edit_1.replace('()','')
    
    edit_3 = edit_2.replace('/',' ') # eksempel: splitte sånt som dette i to -> 'space/locker'
    
    edit_4 = edit_3.replace('Section .','')
    
    # remove spaces at the beginning and at the end of the string
    edit_5 = edit_4.strip()
    
    if edit_5 != '':
        lines.append(edit_5)
    
lines
```

```python
#
# Get the stop words
#

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# Legger til alt som ikkje er til hjelp for å finne dobbeltord
stop_words.update(['.',',','(',')','c',';','..',':','b','c','the','in',"'s",'no','an','etc','where' ])

stop_words
```

```python
#
# Removing stop words
#

stopped_tokens = []

new_lines = []
new_lines_as_text = []

for line in lines:
    
    tokens = word_tokenize(line)
    
    stopped_tokens = stopped_tokens + [token for token in tokens if token.lower() in stop_words]
    
    new_line_in_the_making = [token for token in tokens if token.lower() not in stop_words]
    new_line = " ".join(new_line_in_the_making)
    
    # only add lines that are not blank
    if new_line != '':
        new_lines.append(new_line_in_the_making)
        new_lines_as_text.append(new_line)
    
#print(stopped_tokens)
#print(new_lines)
print(new_lines_as_text)
```

```python
#
# From here we are done changing the text.
# The words will have the same index, even though we have a text with lower case and one normal case text.
#
```

```python
# Get the type of the words

lines_with_word_type = []

for token_line in new_lines:
    tagged_words = nltk.pos_tag(token_line)
    lines_with_word_type.append(tagged_words)
    
lines_with_word_type
```

```python
#
# fjern bøying
#

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''

lines_with_lemmatized_words_and_tag = []
    
for token_line in lines_with_word_type:
    
    lemmatizer = WordNetLemmatizer()
    lemmatized_single_words_with_tag = []

    for (word, pos) in token_line:
        wordnet_pos = get_wordnet_pos(pos)
        if wordnet_pos != '':
            lemmatized_word = lemmatizer.lemmatize(word.lower(), wordnet_pos)
            lemmatized_single_words_with_tag.append((lemmatized_word, pos))
    
    lines_with_lemmatized_words_and_tag.append(lemmatized_single_words_with_tag)
    
lines_with_lemmatized_words_and_tag
```

```python
# each line has an index number
for idx,line in enumerate(lines_with_lemmatized_words_and_tag):
    print(idx)
```

```python
span_count = -1
line_index_with_span_for_lookup = []

for line_idx,line in enumerate(lines_with_lemmatized_words_and_tag):
    
    number_of_words_in_line = len(line)
    first_word_in_sentence_has_index = span_count + 1
    last_word_in_sentence_has_index = first_word_in_sentence_has_index + number_of_words_in_line - 1
    
    # update span counter
    span_count = last_word_in_sentence_has_index
    
    line_index_with_span_for_lookup.append((line_idx,first_word_in_sentence_has_index,last_word_in_sentence_has_index))
    
    print("Line index: " + str(line_idx) + " (" + str(first_word_in_sentence_has_index) + " - " + str(last_word_in_sentence_has_index) + ")")

line_index_with_span_for_lookup
```

```python
span_count = -1
line_index_with_span_for_lookup = []
word_index_for_lookup = []

for line_idx,line in enumerate(lines_with_lemmatized_words_and_tag):
    
    number_of_words_in_line = len(line)
    first_word_in_sentence_has_index = span_count + 1
    last_word_in_sentence_has_index = first_word_in_sentence_has_index + number_of_words_in_line - 1
    
    # update span counter
    span_count = last_word_in_sentence_has_index
    
    line_index_with_span_for_lookup.append((line_idx,first_word_in_sentence_has_index,last_word_in_sentence_has_index))
    
    word_index_count = first_word_in_sentence_has_index
    
    for (word_in_text,pos) in line:
        
        word_with_meta_info = (word_index_count,line_idx,word_in_text,pos)
        word_index_for_lookup.append(word_with_meta_info)
        
        # update counter
        word_index_count = word_index_count + 1
    
    #print("Line index: " + str(line_idx) + " (" + str(first_word_in_sentence_has_index) + " - " + str(last_word_in_sentence_has_index) + ")")

#line_index_with_span_for_lookup
word_index_for_lookup
```

```python
#
# Hypotesis: We must know what to look for.
# Here is what we know.
#

# What are the things we know?
list_of_known_things = ['vessel', 'emergency light', 'pipe system', 'pipe', 'fire extinguisher']

# What are the data properties we know?
list_of_known_data_properties = ['length', 'overall length', 'constructed after', 'constructed before']

# What are the data types we know?
list_of_known_data_types = ['cm', 'metres', 'v', 'mm']

# The big question: Do we find any of these three variables in the same sentence?
# Let's find out.

# Idea: Look for all the things we know.
# Next, in a selected line,
# if a data type is identified, 
# but at known data property is unknown, 
# then make a human identify the unknown data property.
```

```python
#
# Find known data types in line
#

known_data_types_in_line = []

for word_we_know in list_of_known_data_types:
    
    # result
    # word_we_know -> data type example: cm
    # get sentence index
    index_of_lines_with_a_match = []
    
    for (word_index_number,line_index_number,word_in_text,pos) in word_index_for_lookup:

        if word_in_text == word_we_know:
            index_of_lines_with_a_match.append(line_index_number)
            #print(word_in_text + " " + str(line_index_number))
            
    if index_of_lines_with_a_match:
        known_data_types_in_line.append((word_we_know, index_of_lines_with_a_match))

known_data_types_in_line
```

```python
#
# Find known data properties in line
#

known_data_properties_in_line = []

for word_we_know in list_of_known_data_properties:
    
    index_of_lines_with_a_match = []
    
    for (word_index_number,line_index_number,word_in_text,pos) in word_index_for_lookup:

        if word_in_text == word_we_know:
            index_of_lines_with_a_match.append(line_index_number)
            #print(word_in_text + " " + str(line_index_number))
            
    if index_of_lines_with_a_match:
        known_data_properties_in_line.append((word_we_know, index_of_lines_with_a_match))

known_data_properties_in_line
```
