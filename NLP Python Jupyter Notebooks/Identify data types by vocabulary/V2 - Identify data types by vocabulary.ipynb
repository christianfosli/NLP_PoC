```python
#
# Hypotesis: We must know what to look for.
# Here is what we know.
#

# What are the data types we know?
list_of_known_data_types = ['cm', 'metre', 'voltage', 'mm']
```

# single words

```python
#
# Import csv data from script: Create single word with sentence index number and POS
#

import csv

forward_single_words = []

with open('V2 - input - Create single word with sentence index number and POS.csv', newline='\n') as csvfile:
    
    csv_reader = csv.reader(csvfile, delimiter=',')
    
    for row in csv_reader:

        sentence_index_number = row[0]
        word_as_string = row[1]
        word_pos_tag = row[2]
        
        forward_single_words.append((sentence_index_number,word_as_string,word_pos_tag))
        
forward_single_words
```

```python
# Get identified single words with sentence index number

forward_identified_single_words_with_sentence_index_number = []

for (sentence_index_number,word_as_string,word_pos_tag) in forward_single_words:
    
    if word_as_string in list_of_known_data_types:
        forward_identified_single_words_with_sentence_index_number.append((sentence_index_number,word_as_string))
    
forward_identified_single_words_with_sentence_index_number
```

# bigrams

```python
#
# Import csv data from script: Create bigrams with sentence index number and POS
#

import csv

forward_bigrams = []

with open('V2 - input - Create bigrams with sentence index number and POS.csv', newline='\n') as csvfile:
    
    csv_reader = csv.reader(csvfile, delimiter=',')
    
    for row in csv_reader:

        sentence_index = row[0]
        bigram_as_string = row[1]
        word_pos_tag_one = row[2]
        word_pos_tag_two = row[2]
        
        forward_bigrams.append((sentence_index,bigram_as_string,word_pos_tag_one,word_pos_tag_two))
        
forward_bigrams
```

```python
# Get identified bigrams with sentence index number

forward_identified_bigrams_with_sentence_index_number = []

for (sentence_index,bigram_as_string,word_pos_tag_one,word_pos_tag_two) in forward_bigrams:
    
    if bigram_as_string in list_of_known_data_types:
        forward_identified_bigrams_with_sentence_index_number.append((sentence_index,bigram_as_string))
    
forward_identified_bigrams_with_sentence_index_number
```

# Print result to CSV file

```python
import csv

with open("V2 - output - Identify data types by vocabulary.csv", 'w') as f:
    write = csv.writer(f, delimiter=',', lineterminator='\n')
    write.writerows(forward_identified_single_words_with_sentence_index_number)
    write.writerows(forward_identified_bigrams_with_sentence_index_number)
```
