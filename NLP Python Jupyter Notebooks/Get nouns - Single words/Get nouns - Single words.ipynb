```python
#text = "Drip regulation stoves (regulation valve with observation hole) need not have automatic shutoff of the oil supply if the flame will go out due to the oil consumption being limited by an oil regulation valve. The installation instructions shall indicate the maximum permitted difference in height between the oil tank and the regulation valve. An oil pressure reduction valve shall be fitted if satisfactory regulation of the oil supply cannot be achieved through pressure variations. The overflow system shall be capable of receiving the full volume of the tank."

#
# Lastar inn teksten vi jobbar med
#

with open('./Input text - Chapter 3.txt', 'r') as f:
    file_contents = f.read()
    
print(file_contents)
```

```python
#
# Splitte opp linjehopp frå tekst fil.
# Eksempel: Fire, machinery and electric installations\nSection 7.
#

sentences = file_contents.split('\n')
sentences
```

```python
#
# Example to remove something simple without regex by just looping through
#

edited_sentences = []

for line in sentences:
    
    # Bytte eller fjerne teikn
    edited_line = line.replace('/',' ') # eksempel: splitte sånt som dette i to -> 'space/locker'
    edited_sentences.append(edited_line)
    
edited_sentences
```

```python
# Transform list to one text
one_text = " ".join(edited_sentences)
one_text
```

```python
#
# Now we DONT tokenize the lower cased words
# Avoid lowercasing the sentence before word_tokenize because word_tokenize may take cues from capitalization
# tokenized_words = word_tokenize(text_with_lower_case, language='english')
#

nltk.download('punkt')

tokenized_words = word_tokenize(one_text, language='english')
tokenized_words
```

```python
# Get the type of the words

tagged_words = nltk.pos_tag(tokenized_words)
tagged_words
```

```python
from nltk.probability import FreqDist

frequency_distribution = FreqDist(tagged_words)
frequency_distribution
```

```python
# List ut det dei forskjellige førekomstane i teksten

fd_tagged = FreqDist(tag for (word, tag) in tagged_words)
fd_tagged.most_common(10)
```

```python
# List ut alle NN (som er det same som nouns / substantiv)

# this will only list the nouns
#nouns = [word for (word, pos) in frequency_distribution if pos == 'NN']

# this will list the words with their types
nouns = [(word, pos) for (word, pos) in frequency_distribution if pos == 'NN']

nouns
```

```python
# fjern bøying

# Ein lyt spessifisere kva ein bøyger. Derom ein vil bøyge noko anna enn substantiv kan ein gjere det sånn:
# https://stackoverflow.com/questions/15586721/wordnet-lemmatization-and-pos-tagging-in-python?noredirect=1&lq=1

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

lemmatized_nouns = []

for (noun, pos) in nouns:
    lemmatized_noun = lemmatizer.lemmatize(noun.lower(), wordnet.NOUN)
    lemmatized_nouns.append((lemmatized_noun, pos))

lemmatized_nouns
```

```python
import pandas as pd 
pd.DataFrame(lemmatized_nouns).to_csv("output - Get nouns - Single words.csv", header=None, index=None)
```
