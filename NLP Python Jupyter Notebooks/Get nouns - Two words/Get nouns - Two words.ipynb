```python
import nltk
print(nltk.__version__)
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
```

```python
#
# Download
#

nltk.download('stopwords')
```

```python
#
# Lastar inn teksten vi jobbar med
#

with open('./Input text - Chapter 3.txt', 'r') as f:
    file_contents = f.read()
    
print(file_contents)
```

```python
#
# Splitte opp linjehopp frå tekst fil.
# Eksempel: Fire, machinery and electric installations\nSection 7.
#

sentences = file_contents.split('\n')
sentences
```

```python
# Transform list to one text
one_text = " ".join(sentences)
one_text
```

```python
# Splitte opp tekst etter setningar.
# Det gir ikkje meining å smelete setningar saman når ein leiter etter ord som høyrer saman.
# Vi leitar derfor etter ord som høyrer saman innanfor ein setning.

from nltk.tokenize.punkt import PunktSentenceTokenizer

pst = PunktSentenceTokenizer()

sent_tokens = pst.tokenize(one_text)

sent_tokens
```

```python
#
# Clean the lines up
#

import re

lines = []

for line in sent_tokens:
    
    # remove all numbers
    edit_1 = re.sub(r"\d","", line)
    
    edit_2 = edit_1.replace('()','')
    
    edit_3 = edit_2.replace('/',' ') # eksempel: splitte sånt som dette i to -> 'space/locker'
    
    edit_4 = edit_3.replace('Section .','')
    
    # remove spaces at the beginning and at the end of the string
    edit_5 = edit_4.strip()
    
    if edit_5 != '':
        lines.append(edit_5)
    
lines
```

```python
#
# Get the stop words
#

stop_words = set(stopwords.words('english'))

# Legger til alt som ikkje er til hjelp for å finne dobbeltord
stop_words.update(['.',',','(',')','°c','c',';','..',':','b','c','the','mm','cm','in',"'s",'no','an','etc','where' ])

stop_words
```

```python
stopped_tokens = []

new_lines = []

for line in lines:
    
    tokens = word_tokenize(line)
    
    stopped_tokens = stopped_tokens + [token for token in tokens if token.lower() in stop_words]
    
    new_line_in_the_making = [token for token in tokens if token.lower() not in stop_words]
    new_line = " ".join(new_line_in_the_making)
    
    if new_line != '':
        new_lines.append(new_line_in_the_making)
    
#print(stopped_tokens)
print(new_lines)
```

```python
# Get the type of the words

lines_with_word_type = []

for token_line in new_lines:
    tagged_words = nltk.pos_tag(token_line)
    lines_with_word_type.append(tagged_words)
    
lines_with_word_type
```

```python
# fjern bøying

# Ein lyt spessifisere kva ein bøyger. Derom ein vil bøyge noko anna enn substantiv kan ein ghere det sånn:
# https://stackoverflow.com/questions/15586721/wordnet-lemmatization-and-pos-tagging-in-python?noredirect=1&lq=1

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''

lines_with_lemmatized_words_and_tag = []
    
for token_line in lines_with_word_type:
    
    lemmatizer = WordNetLemmatizer()
    lemmatized_single_words_with_tag = []

    for (word, pos) in token_line:
        wordnet_pos = get_wordnet_pos(pos)
        if wordnet_pos != '':
            lemmatized_word = lemmatizer.lemmatize(word.lower(), wordnet_pos)
            lemmatized_single_words_with_tag.append((lemmatized_word, pos))
    
    lines_with_lemmatized_words_and_tag.append(lemmatized_single_words_with_tag)
    
lines_with_lemmatized_words_and_tag
```

```python
lines = []

for line in lines_with_lemmatized_words_and_tag:
    
    lemmatized_single_words_without_type = []

    for (word, pos) in line:
        
        # skip duplicate words
        if lemmatized_single_words_without_type[-1:] == []:
            lemmatized_single_words_without_type.append(word)
        elif lemmatized_single_words_without_type[-1] != word:
            lemmatized_single_words_without_type.append(word)
        
    lines.append(" ".join(lemmatized_single_words_without_type))
    
lines
```

```python
from nltk import word_tokenize 
from nltk.util import ngrams

nltk_bigrams = []

for line in lines:
    word_tokens = word_tokenize(line)
    bigram = list(ngrams(word_tokens, 2))
    #print(bigram)
    nltk_bigrams.append(bigram)
    
nltk_bigrams
```

```python
# slå saman alle bigrams til ei liste
# dette går fint etter at bigrams'a er laga

bigram_list = []

for line in nltk_bigrams:
    for bigram in line:
        bigram_list.append(bigram)
        
bigram_list
```

```python
from nltk.collocations import BigramAssocMeasures
from nltk.collocations import BigramCollocationFinder

bigram_measures = BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(lemmatized_single_words_without_type)

# return the top n-grams with the highest PMI
#finder.nbest(bigram_measures.likelihood_ratio, 100) # talet er ein limit
finder.nbest(bigram_measures.likelihood_ratio, 1000)
```

```python
# The simplest method is to rank the most frequent bigrams.
# Source: https://medium.com/@nicharuch/collocations-identifying-phrases-that-act-like-individual-words-in-nlp-f58a93a2f84a

import pandas as pd 

#bigrams
bigram_freq = finder.ngram_fd.items()
bigramFreqTable = pd.DataFrame(list(bigram_freq), columns=['bigram','freq']).sort_values(by='freq', ascending=False)
bigramFreqTable

# However, a common issue with this is adjacent spaces, stop words, articles, prepositions or pronouns are common and are not meaningful.
```

```python
# To fix this, we filter out for collocations not containing stop words and filter for only the following structures:
# Bigrams: (Noun, Noun), (Adjective, Noun)
# This is a common structure used in literature and generally works well.

#function to filter for ADJ/NN bigrams
def rightTypes(ngram):
    
    if '-pron-' in ngram or 't' in ngram:
        return False
    
    for word in ngram:
        if word in stop_words or word.isspace():
            return False
        
    acceptable_types = ('JJ', 'JJR', 'JJS', 'NN', 'NNS', 'NNP', 'NNPS')
    second_type = ('NN', 'NNS', 'NNP', 'NNPS')
    tags = nltk.pos_tag(ngram)
    
    if tags[0][1] in acceptable_types and tags[1][1] in second_type:
        return True
    else:
        return False

#filter bigrams
filtered_bi = bigramFreqTable[bigramFreqTable.bigram.map(lambda x: rightTypes(x))]
filtered_bi
```

```python
pd.DataFrame(filtered_bi).to_csv("output - Get nouns - Two words.csv", header=None, index=None)
```
