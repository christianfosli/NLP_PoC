```python
import nltk
nltk_text = nltk.Text(word_tokens)
match = nltk_text.concordance('Section')
```

```python
nltk_concordance_index = nltk.ConcordanceIndex(word_tokens)
```

```python
# get the index of a searched word
print([offset for offset in nltk_concordance_index.offsets("Section")])
```

```python
# get the word before a searched word
print([word_tokens[offset - 1] for offset in nltk_concordance_index.offsets("Section")])
```

```python
# get the word after a searched word
# Result: this is the Section numbers that we are looking for
print([word_tokens[offset + 1] for offset in nltk_concordance_index.offsets("Section")])
```

```python
#
# Liste med Section og index
#

section_overview = []

for offset in nltk_concordance_index.offsets("Section"):
    section_overview.append([
        offset, # index
        word_tokens[offset], 
        word_tokens[offset + 1], 
        word_tokens[offset + 2]
    ])
    
section_overview
```

```python
# Fuction that find section of an index number

def get_section_from_word_index_number(word_index_number):
    
    last_section_memory = []
    
    for section in section_overview:
        if section[0] < word_index_number:
            last_section_memory.append(section[2])
        else:
            break
            
    # returns the last section or empty list
    result_in_a_array = last_section_memory[-1:]
    
    # return as string
    return result_in_a_array[0]

get_section_from_word_index_number(1000)
```

```python
#
# Find something
#

for offset in nltk_concordance_index.offsets("than"):
    print([
        offset, # index
        get_section_from_word_index_number(offset),
        word_tokens[offset - 2], 
        word_tokens[offset - 1], 
        word_tokens[offset], 
        word_tokens[offset + 1], 
        word_tokens[offset + 2]
    ])
```

```python
def is_int(n):
    try:
        float_n = float(n)
        int_n = int(float_n)
    except ValueError:
        return False
    else:
        return float_n == int_n

# test
print(is_int('test'))
print(is_int('4'))
print(is_int('4.5'))
```

```python
def is_float(n):
    try:
        float_n = float(n)
    except ValueError:
        return False
    else:
        return True
    
# test
print(is_float('test'))
print(is_float('4'))
print(is_float('4.5'))
```

```python
#
# Find "than" + floats
#

result_for_csv_file_than = []

for offset in nltk_concordance_index.offsets("than"):

    if is_float(word_tokens[offset + 1]):
        print([
            offset, # index
            get_section_from_word_index_number(offset),
            word_tokens[offset - 2], 
            word_tokens[offset - 1], 
            word_tokens[offset], 
            word_tokens[offset + 1], 
            word_tokens[offset + 2]
        ])
        
        paragraph_number = "Section " + get_section_from_word_index_number(offset)
        relation = word_tokens[offset - 1] + " " + word_tokens[offset]
        value_1 = word_tokens[offset + 1]
        value_2 = word_tokens[offset + 2]
        
        if value_2 == "meters" or value_2 == "metres": #misspelling
            result_for_csv_file_than.append([
                paragraph_number,
                relation,
                value_1,
                "meters"
                ])
        
list(result_for_csv_file_than)
```

```python
#
# Find "to" + floats
#

result_for_csv_file_to = []

for offset in nltk_concordance_index.offsets("to"):

    if is_float(word_tokens[offset + 1]):
        
        print([
            offset, # index
            get_section_from_word_index_number(offset),
            word_tokens[offset - 1], 
            word_tokens[offset], 
            word_tokens[offset + 1], 
            word_tokens[offset + 2]
        ])
        
        paragraph_number = "Section " + get_section_from_word_index_number(offset)
        value_2 = word_tokens[offset + 1]
        
        if is_float(word_tokens[offset - 1]):
            value_1 = word_tokens[offset - 1]
            relation = word_tokens[offset]
        else:
            value_1 = ''
            relation = word_tokens[offset - 1] + " " + word_tokens[offset]
        
        if word_tokens[offset + 2] == "meters" or word_tokens[offset + 2] == "metres": #misspelling
            result_for_csv_file_to.append([
                paragraph_number,
                relation,
                value_1,
                value_2,
                "meters"
                ])
        
list(result_for_csv_file_to)
```

```python
import csv
wtr = csv.writer(open ("output - Get scope - Vessel length overall.csv", 'w'), delimiter=',', lineterminator='\n')
for x in result_for_csv_file_than : wtr.writerow ([x])
for x in result_for_csv_file_to : wtr.writerow ([x])
```
