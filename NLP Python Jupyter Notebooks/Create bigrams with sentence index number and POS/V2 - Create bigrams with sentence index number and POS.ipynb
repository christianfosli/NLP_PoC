```python
#
# Lastar inn teksten vi jobbar med
#

with open('./V2 - input - Chapter 3 - Text from word document with correct formating.txt', 'r') as f:

    file_lines = f.readlines() 
    
print(file_lines)
```

```python
# Does the input string contain a digit?
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def hasSlash(inputString):
    return any(char == '/' for char in inputString)
```

```python
#
# Make line ready for POS creation
#

from nltk.tokenize import word_tokenize

lines_ready_for_pos_creation = []

for line_index_number, line_text in enumerate(file_lines):

    tokens = word_tokenize(line_text)
    
    lines_ready_for_pos_creation.append(tokens)
    
lines_ready_for_pos_creation
```

```python
# Get the type of the words

import nltk

forward_sentences_in_word_list_with_pos = []

for token_line in lines_ready_for_pos_creation:
    tagged_words = nltk.pos_tag(token_line)
    forward_sentences_in_word_list_with_pos.append(tagged_words)
    
forward_sentences_in_word_list_with_pos
```

```python
#
# Print bigram on all text
# Chalange: Can we keep the POS through the bigram process? This can be very useful! When lemmatizing..
#

from nltk import ngrams

forward_bigrams_with_pos_preserved = []

for sentence_index_number,sentence_as_a_string in enumerate(file_lines):
    
    tokenized_sentence = nltk.word_tokenize(sentence_as_a_string)
    sentence_pos_tagged_by_word_list = nltk.pos_tag(tokenized_sentence)
    
    sentence_as_pos = []
    
    for word_index_number,tokenized_word in enumerate(tokenized_sentence):
        #print(tokenized_sentence[word_index_number]) # the word
        #print(sentence_pos_tagged_by_word_list[word_index_number][1]) # POS of the word
        sentence_as_pos.append(sentence_pos_tagged_by_word_list[word_index_number][1])
    
    # Now these two are the same, but one is the POS
    # Idea: We run bigram on both to keep the POS
    #print(tokenized_sentence)
    #print(sentence_as_pos)
    
    bigram_on_tokenized_sentence = list(ngrams(tokenized_sentence, 2))
    bigram_on_sentence_as_pos = list(ngrams(sentence_as_pos, 2))
    #print(bigram_on_tokenized_sentence)
    #print(bigram_on_sentence_as_pos)
    
    # We did it!!
    
    # Now we merge the results to one list
    bigram_on_tokenized_sentence_with_pos_preserved = (bigram_on_tokenized_sentence,bigram_on_sentence_as_pos)
    forward_bigrams_with_pos_preserved.append(bigram_on_tokenized_sentence_with_pos_preserved)
    
forward_bigrams_with_pos_preserved
```

```python
#
# Example that we have the sentence_index_number, the bigram, with POS
#

for sentence_index,sentence in enumerate(forward_bigrams_with_pos_preserved):
    bigram_on_tokenized_sentence = sentence[0]
    bigram_on_sentence_as_pos = sentence[1]
    
    for word_index,word in enumerate(bigram_on_tokenized_sentence):
        print("sentence_index: " + str(sentence_index) + ": " + bigram_on_tokenized_sentence[word_index][0] + " " + bigram_on_tokenized_sentence[word_index][1])
        print(bigram_on_sentence_as_pos[word_index])
```

```python
#
# Get the stop words
#

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

forward_english_stopwords = set(stopwords.words('english'))

# Legger til alt som ikkje er til hjelp for å finne dobbeltord
forward_english_stopwords.update(['.',',','(',')','c',';','..',':','b','c','the','in',"'s",'no','an','etc','where','-'])

# Removing the months
forward_english_stopwords.update(['january','february','march','april','may','june','july','august','september','october','november','december'])

forward_english_stopwords
```

```python
#
# Print bigram with no stopwords
#

from nltk import ngrams

for sentence_index_number,sentence_as_a_string in enumerate(file_lines):
    token = nltk.word_tokenize(sentence_as_a_string)
    bigram_list = list(ngrams(token, 2))
    
    for bigram in bigram_list:
        
        if hasNumbers(bigram[0]) or hasNumbers(bigram[1]):
            continue # skip if containing numbers
        elif hasSlash(bigram[0]) or hasSlash(bigram[1]):
            continue # skip if containing /
        
        bigram_in_a_list = [bigram[0],bigram[1]]
        
        bigram_without_stopword = [bigram_split for bigram_split in bigram_in_a_list if bigram_split.lower() not in forward_english_stopwords]
        
        if len(bigram_without_stopword) == 2:
            print(bigram_without_stopword)
        
```

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
#
# Lemmatize bigram with no stopwords AND keep POS
#

forward_result_to_csv = []

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')

for sentence_index,sentence in enumerate(forward_bigrams_with_pos_preserved):
    bigram_on_tokenized_sentence = sentence[0]
    bigram_on_sentence_as_pos = sentence[1]
    
    for bigram_index_number,bigram in enumerate(bigram_on_tokenized_sentence):
        
        #print("sentence_index: " + 
        # str(sentence_index) + ": " + 
        # bigram_on_tokenized_sentence[word_index][0] + 
        # " " + bigram_on_tokenized_sentence[word_index][1])
        
        #print(bigram_on_sentence_as_pos[word_index])

        if hasNumbers(bigram[0]) or hasNumbers(bigram[1]):
            continue # skip if containing numbers
        elif hasSlash(bigram[0]) or hasSlash(bigram[1]):
            continue # skip if containing /
        
        bigram_in_a_list = [bigram[0],bigram[1]]
        
        bigram_without_stopword = [bigram_split for bigram_split in bigram_in_a_list if bigram_split.lower() not in forward_english_stopwords]
        
        if len(bigram_without_stopword) != 2:
            continue
            
        #print(bigram[0] + " " + bigram[1] + " (" + bigram_on_sentence_as_pos[bigram_index_number][0] + " " + bigram_on_sentence_as_pos[bigram_index_number][1] + ")")
        #print(bigram_without_stopword)
        #print(bigram_on_sentence_as_pos[bigram_index_number])
        
        #
        # Time to lemmatize the words.
        # This is what we need the POS for!
        #
        
        wordnet_pos_one = get_wordnet_pos(bigram_on_sentence_as_pos[bigram_index_number][0])
        wordnet_pos_two = get_wordnet_pos(bigram_on_sentence_as_pos[bigram_index_number][1])
        
        lemmatizer = WordNetLemmatizer()

        word_one = bigram[0].lower()
        word_two = bigram[1].lower()
        
        if wordnet_pos_one != '':
            word_one = lemmatizer.lemmatize(bigram[0].lower(), wordnet_pos_one)
        
        if wordnet_pos_two != '':
            word_two = lemmatizer.lemmatize(bigram[1].lower(), wordnet_pos_two)
        
        #print(str(sentence_index) + ": " + word_one + " " + word_two)
        
        # the output:
        # - sentence_index
        bigram_as_string = word_one + " " + word_two # - bigram
        # - the POS taggs
        
        forward_result_to_csv.append((str(sentence_index),bigram_as_string,bigram_on_sentence_as_pos[bigram_index_number][0],bigram_on_sentence_as_pos[bigram_index_number][1]))
        
forward_result_to_csv
```

```python
import csv

with open("V2 - output - Create bigrams with sentence index number and POS.csv", 'w') as f:
    write = csv.writer(f, delimiter=',', lineterminator='\n')
    write.writerows(forward_result_to_csv)
```
