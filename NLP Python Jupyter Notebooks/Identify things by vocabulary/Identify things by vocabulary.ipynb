# bigrams

```python
#
# Import csv data from script: Create bigrams with sentence index number and POS
#

import csv

forward_bigrams = []

with open('input - Create bigrams with sentence index number and POS.csv', newline='\n') as csvfile:
    
    csv_reader = csv.reader(csvfile, delimiter=',')
    
    for row in csv_reader:

        sentence_index = row[0]
        bigram_as_string = row[1]
        word_pos_tag_one = row[2]
        word_pos_tag_two = row[2]
        
        forward_bigrams.append((sentence_index,bigram_as_string,word_pos_tag_one,word_pos_tag_two))
        
forward_bigrams
```

```python
# Get identified bigrams with sentence index number

forward_identified_bigrams_with_sentence_index_number = []

for (sentence_index,bigram_as_string,word_pos_tag_one,word_pos_tag_two) in forward_bigrams:
    
    if bigram_as_string in list_of_known_things:
        forward_identified_bigrams_with_sentence_index_number.append((sentence_index,bigram_as_string))
    
forward_identified_bigrams_with_sentence_index_number
```

# Print result to CSV file

```python
import csv

with open("output - Identified bigram things by vocabulary.csv", 'w') as f:
    write = csv.writer(f, delimiter=',', lineterminator='\n')
    write.writerows(forward_identified_bigrams_with_sentence_index_number)
```
