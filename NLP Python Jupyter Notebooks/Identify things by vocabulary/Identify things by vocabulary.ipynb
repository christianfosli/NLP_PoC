```python
#
# Import csv data from script: Create bigrams with sentence index number and POS
#

import csv

forward_bigrams = []

with open('input - Create bigrams with sentence index number and POS.csv', newline='\n') as csvfile:
    
    csv_reader = csv.reader(csvfile, delimiter=',')
    
    for row in csv_reader:

        sentence_index = row[0]
        bigram_as_string = row[1]
        word_pos_tag_one = row[2]
        word_pos_tag_two = row[2]
        
        forward_bigrams.append((sentence_index,bigram_as_string,word_pos_tag_one,word_pos_tag_two))
        
forward_bigrams
```

```python
#
# Find known data types in line
#

known_data_types_in_line = []

for word_we_know in list_of_known_data_types:
    
    # result
    # word_we_know -> data type example: cm
    # get sentence index
    index_of_lines_with_a_match = []
    
    for (word_index_number,line_index_number,word_in_text,pos) in word_index_for_lookup:

        if word_in_text == word_we_know:
            index_of_lines_with_a_match.append(line_index_number)
            #print(word_in_text + " " + str(line_index_number))
            
    if index_of_lines_with_a_match:
        known_data_types_in_line.append((word_we_know, index_of_lines_with_a_match))

known_data_types_in_line
```

```python
#
# Find known data properties in line
#

known_data_properties_in_line = []

for word_we_know in list_of_known_data_properties:
    
    index_of_lines_with_a_match = []
    
    for (word_index_number,line_index_number,word_in_text,pos) in word_index_for_lookup:

        if word_in_text == word_we_know:
            index_of_lines_with_a_match.append(line_index_number)
            #print(word_in_text + " " + str(line_index_number))
            
    if index_of_lines_with_a_match:
        known_data_properties_in_line.append((word_we_know, index_of_lines_with_a_match))

known_data_properties_in_line
```

```python
# List the findings

findings = []

for (word_we_know, index_of_lines_with_a_match) in known_data_types_in_line:
    for index_of_line in index_of_lines_with_a_match:
        findings.append((index_of_line,word_we_know))

sortedList = sorted(findings, key = lambda y: (y[0]))
sortedList
```

```python
# Create dictionary with findings of known_data_types_in_line
# The dictionary key is the index number of the line

from collections import defaultdict

dictionary_with_known_data_types_in_a_sentence = defaultdict(int)

for (word_we_know, index_of_lines_with_a_match) in known_data_types_in_line:
    
    for index_of_line in index_of_lines_with_a_match:
        
        if index_of_line not in dictionary_with_known_data_types_in_a_sentence:
            
            print('First detection in sentence ' + str(index_of_line) + " - " + word_we_know)
            
            # Add to dictionary
            dictionary_with_known_data_types_in_a_sentence[index_of_line] = [word_we_know]
            
        else: # a detection in sentence is already added
            
            print('New detection in sentence ' + str(index_of_line) + " - " + word_we_know)
            
            # Add if it is a new word
            if word_we_know not in dictionary_with_known_data_types_in_a_sentence[index_of_line]:
                dictionary_with_known_data_types_in_a_sentence[index_of_line].append(word_we_know)
                
dictionary_with_known_data_types_in_a_sentence
```

```python
# Create dictionary with findings of known_data_properties_in_line
# The dictionary key is the index number of the line

from collections import defaultdict

dictionary_with_known_data_properties_in_a_sentence = defaultdict(int)

for (word_we_know, index_of_lines_with_a_match) in known_data_properties_in_line:
    
    for index_of_line in index_of_lines_with_a_match:
        
        if index_of_line not in dictionary_with_known_data_properties_in_a_sentence:
            
            print('First detection in sentence ' + str(index_of_line) + " - " + word_we_know)
            
            # Add to dictionary
            dictionary_with_known_data_properties_in_a_sentence[index_of_line] = [word_we_know]
            
        else: # a detection in sentence is already added
            
            print('New detection in sentence ' + str(index_of_line) + " - " + word_we_know)
            
            # Add if it is a new word
            if word_we_know not in dictionary_with_known_data_properties_in_a_sentence[index_of_line]:
                dictionary_with_known_data_properties_in_a_sentence[index_of_line].append(word_we_know)
                
dictionary_with_known_data_properties_in_a_sentence
```

```python
# Merge dictionary_with_known_data_types_in_a_sentence and dictionary_with_known_data_properties_in_a_sentence

dictionary_with_known_data_types_and_properties_in_a_sentence = defaultdict(int)

# Load data type line numbers
for line_number in dictionary_with_known_data_types_in_a_sentence.keys():
    dictionary_with_known_data_types_and_properties_in_a_sentence[line_number] = ([],[])

# Load data property line numbers
for line_number in dictionary_with_known_data_properties_in_a_sentence.keys():
    dictionary_with_known_data_types_and_properties_in_a_sentence[line_number] = ([],[])

# Load data
for line_number in dictionary_with_known_data_types_and_properties_in_a_sentence.keys():
    data_types = dictionary_with_known_data_types_in_a_sentence[line_number]
    data_properties = dictionary_with_known_data_properties_in_a_sentence[line_number]    
    dictionary_with_known_data_types_and_properties_in_a_sentence[line_number] = (data_types,data_properties)

# Result comment: easy to see what we have.
# A human must look at the sentences with 0 values
    
dictionary_with_known_data_types_and_properties_in_a_sentence
```

```python
# Focus on sentences with findings on both type AND property

dictionary_with_match_on_type_and_property = defaultdict(int)
dictionary_with_match_on_type_or_property = defaultdict(int)

for line_number in dictionary_with_known_data_types_and_properties_in_a_sentence.keys():
    data_types = dictionary_with_known_data_types_and_properties_in_a_sentence[line_number][0]
    data_properties = dictionary_with_known_data_types_and_properties_in_a_sentence[line_number][1]
    
    if data_types == 0 or data_properties == 0:
        dictionary_with_match_on_type_or_property[line_number] = (data_types,data_properties)
    else:
        dictionary_with_match_on_type_and_property[line_number] = (data_types,data_properties)

dictionary_with_match_on_type_and_property
#dictionary_with_match_on_type_or_property
```

```python
#
# Here we have found sentences with something, but type OR property is missing.
# This is a good place for a human to look for sentences with unknown words.
#

dictionary_with_match_on_type_and_property = defaultdict(int)
dictionary_with_match_on_type_or_property = defaultdict(int)

for line_number in dictionary_with_known_data_types_and_properties_in_a_sentence.keys():
    data_types = dictionary_with_known_data_types_and_properties_in_a_sentence[line_number][0]
    data_properties = dictionary_with_known_data_types_and_properties_in_a_sentence[line_number][1]
    
    if data_types == 0 or data_properties == 0:
        dictionary_with_match_on_type_or_property[line_number] = (data_types,data_properties)
    else:
        dictionary_with_match_on_type_and_property[line_number] = (data_types,data_properties)

#dictionary_with_match_on_type_and_property
dictionary_with_match_on_type_or_property
```

# Identify known and unknown things in sentences

```python
#
# Find known words in line
#
# Now we have identified lines with known types and properties.
# Next up, can we identify the thing we are talking about?
#
# Idea: Remove the identified words from the sentences before continuing to searching in them

selected_line_in_a_list_of_words = []

for (word_index_number,line_index_number,word_in_text,pos) in word_index_for_lookup:
    if line_index_number == 105:
        selected_line_in_a_list_of_words.append((word_index_number,line_index_number,word_in_text,pos))

line_with_findings_removed = [(word_index_number,line_index_number,word_in_text,pos) for (word_index_number,line_index_number,word_in_text,pos) in selected_line_in_a_list_of_words if word_in_text not in list_of_known_data_types]
line_with_findings_removed
```

```python
#
# Find sentences where we are going to look for something (the subject in the RDf triple)
#
# Now we have identified lines with known types and properties.
# Next up, can we identify the thing we are talking about?
#
# Idea: Remove the identified words from the sentences before continuing to searching in them

sentences_without_data_type_and_data_property = []

for line_number in dictionary_with_match_on_type_and_property.keys():
    for (word_index_number,line_index_number,word_in_text,pos) in word_index_for_lookup:
        
        # only add words from sentence that is not data_type or data_property
        data_types = dictionary_with_match_on_type_and_property[line_number][0]
        data_properties = dictionary_with_match_on_type_and_property[line_number][1]

        if line_index_number == line_number and word_in_text not in data_types and word_in_text not in data_properties:
            sentences_without_data_type_and_data_property.append((word_index_number,line_index_number,word_in_text,pos))

sentences_without_data_type_and_data_property
```
