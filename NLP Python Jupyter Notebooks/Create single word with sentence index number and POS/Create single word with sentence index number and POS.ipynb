```python
#
# Lastar inn teksten vi jobbar med
#

with open('./Input text - Chapter 3.txt', 'r') as f:
    file_contents = f.read()
    
print(file_contents)
```

```python
#
# Splitte opp linjehopp frå tekst fil.
# Eksempel: Fire, machinery and electric installations\nSection 7.
#

forward_file_content_split = file_contents.split('\n')
forward_file_content_split
```

```python
#
# Transform list to one text
#

forward_one_text = " ".join(forward_file_content_split)
forward_one_text
```

```python
# Does the input string contain a digit?
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def hasSlash(inputString):
    return any(char == '/' for char in inputString)
```

```python
#
# Make selected words more explicit.
# Normalize for the machine.
#

from nltk.tokenize import word_tokenize

word_tokens = word_tokenize(forward_one_text, language="english")

forward_word_tokens = []

for word_tokens_index_number,tokenized_word in enumerate(word_tokens):
    
    if tokenized_word.lower() == 'v':
        tokenized_word = word_tokenize('voltage', language="english")[0]
        
    elif tokenized_word.lower() == 'm':
        tokenized_word = word_tokenize('metre', language="english")[0]
        
    elif tokenized_word.lower() == 'cm':
        tokenized_word = word_tokenize('centimetre', language="english")[0]
        
    elif tokenized_word.lower() == 'mm':
        tokenized_word = word_tokenize('millimetre', language="english")[0]
        
    elif tokenized_word.lower() == 'kg':
        tokenized_word = word_tokenize('kilogram', language="english")[0]
        
    elif tokenized_word == 'Section' and word_tokens_index_number > 0:
        
        previous_word = forward_word_tokens[-1]
        
        next_word_index_number = word_tokens_index_number + 1
        next_word = word_tokens[next_word_index_number]
        
        next_next_word_index_number = word_tokens_index_number + 2
        next_next_word = word_tokens[next_next_word_index_number]
        
        if previous_word != '.' and hasNumbers(next_word) and next_next_word == '.':
            # insert . (dot) before the word
            dot = word_tokenize('.', language="english")[0]
            forward_word_tokens.append(dot)
    
    forward_word_tokens.append(tokenized_word)

forward_word_tokens
```

# After this point, dont edit the source text, as we need a stable index number on it

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer
forward_normalized_text = TreebankWordDetokenizer().detokenize(forward_word_tokens)
forward_normalized_text
```

```python
#
# Create sentences
#

from nltk.tokenize.punkt import PunktSentenceTokenizer

pst = PunktSentenceTokenizer()

forward_sentences = pst.sentences_from_tokens(forward_word_tokens)
list(forward_sentences)
```

```python
# Splitte opp tekst etter setningar.
# Vi leitar derfor etter ord som høyrer saman innanfor ein setning.

forward_sent_tokens = pst.tokenize(forward_normalized_text)
forward_sent_tokens
```

```python
#
# Each line has a sentence index number.
# It is important that we keep these sentence index number.
# It will be the key to get a result in the end.
#

for sentence_index_number,sentence_as_a_string in enumerate(forward_sent_tokens):
    print("--- sentence_index_number: " + str(sentence_index_number) + " ---")
    print(sentence_as_a_string)
```

```python
#
# Make line ready for POS creation
#

lines_ready_for_pos_creation = []

for sentence_index_number,sentence_as_a_string in enumerate(forward_sent_tokens):
    
    tokens = word_tokenize(sentence_as_a_string)
    
    lines_ready_for_pos_creation.append(tokens)
    
lines_ready_for_pos_creation
```

```python
# Get the type of the words

import nltk

forward_sentences_in_word_list_with_pos = []

for token_line in lines_ready_for_pos_creation:
    tagged_words = nltk.pos_tag(token_line)
    forward_sentences_in_word_list_with_pos.append(tagged_words)
    
forward_sentences_in_word_list_with_pos
```

```python
#
# Get the stop words
#

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

forward_english_stopwords = set(stopwords.words('english'))

# Legger til alt som ikkje er til hjelp for å finne dobbeltord
forward_english_stopwords.update(['.',',','(',')','c',';','..',':','b','c','the','in',"'s",'no','an','etc','where','-'])

forward_english_stopwords
```

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
#
# Lemmatize single words with no stopwords AND keep POS
#

forward_result_to_csv = []

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')

for sentence_index_number,sentence in enumerate(forward_sentences_in_word_list_with_pos):
    
    for word in sentence:
        
        #print(word)
        
        word_as_lower = word[0].lower()
        pos_tag = word[1]
        
        #
        # lemmatizing
        #
        
        wordnet_pos_tag = get_wordnet_pos(pos_tag)
        
        lemmatizer = WordNetLemmatizer()
        
        if wordnet_pos_tag == '':
            continue
            
        word_as_lower = lemmatizer.lemmatize(word_as_lower, wordnet_pos_tag)
        
        #
        # stopwords
        #
        
        if word_as_lower in forward_english_stopwords:
            continue
        
        #if word_as_lower == 'metre':
            #forward_result_to_csv.append((sentence_index_number,word[0],word_as_lower,pos_tag))
            
        forward_result_to_csv.append((sentence_index_number,word_as_lower,pos_tag))
        
forward_result_to_csv
```

# Write output to CSV

```python
import csv

with open("output - Create single word with sentence index number and POS.csv", 'w') as f:
    write = csv.writer(f, delimiter=',', lineterminator='\n')
    write.writerows(forward_result_to_csv)
```
